```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generate data centered around 0
x = np.linspace(-5, 5)
y = 2 * x + 1

# 2. Setup the figure
fig, ax = plt.subplots(figsize=(8, 5))

# Move the left and bottom spines to the center (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Hide the top and right spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 3. Plot the function
ax.plot(x, y, label=r'$y = 2x + 1$', color='crimson', linewidth=2)

# Add a grid and set axis limits to keep it symmetrical
ax.grid(True, linestyle=':', alpha=0.6)
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5]) # Adjusted for y = 2(10)+1

# Place the legend in a clear spot
ax.legend(loc='upper left')

intercepts_x = [0, -0.5]
intercepts_y = [1, 0]
ax.scatter(intercepts_x, intercepts_y, color='red', s=50, zorder=2)
# Annotate the points for clarity
ax.annotate('(0, 1)', (0.3, 1.5), fontsize=10, color='red')
ax.annotate('(-0.5, 0)', (-2.5, -1.5), fontsize=10, color='red')

# 4. Save to the Antigravity workspace
plt.savefig("centered_linear_plot.png", dpi=300)
plt.show()
```
