# Neural Net
Utilizar el dataset mnist para entender cómo funciona una red neuronal
El objetivo será clasificar observaciones de imágenes de 28x28 pixeles para determinar qué dígito del 0 al 9 corresopnden

```python
#dependencies
import matplotlib.pyplot as plt #librería para hacer gráficos
import numpy as np #librería de matemática
import tensorflow as tf #herramienta para problemas de ML
import tensorflow_datasets as tfds #librería de datasets predeterminados de tensorflow
```

```python
#1. Importar el dataset 'mnist', con la información para entrenar.
ds = tfds.load('mnist', split='train', as_supervised=True) # as_supervised nos da una tupla (feature,label)
```

Para cada elemento del dataset tendremos
- Una imagen de 28x28 pixeles, donde cada pixel tiene un valor que determina la "opacidad". 
- Una etiqueta que nos indica qué dígito del 0 al 9 representa esa imagen 

```python
#2. Analizamos la información de una observación
one = ds.take(1)
for image, label in one:
    print(f"La observación tiene {image.shape} dimensiones")
    print(f"La etiqueta dice que esta observación es el número  {label}")

for image, label in tfds.as_numpy(one): #tfds.an_numpy transforma un tensor en un array
  print(type(image), type(label), label)

ds, info = tfds.load('mnist', split='train', with_info=True) #info tiene la metadata de la observación
tfds.as_dataframe(ds.take(4), info)
```

```python
features = info.features #features = parámetros
print(features)
print(features['label'].num_classes) #cantidad de etiquetas distintas
print(features['label'].names) #cada etiqueta distinta
print(features['image'].shape) # las dimensiones de cada imagen
print(features['image'].np_dtype)
splits = info.splits
print(splits) # las particiones en entrenamiento / test
```
